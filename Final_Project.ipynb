```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample HR data
data = {
    'Gender': ['Male', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'PerformanceRating': [3, 4, 3, 2, 4, 3, 4, 2, 3, 4]
}

df = pd.DataFrame(data)

# Group data by Gender and calculate average performance rating
grouped_data = df.groupby('Gender')['PerformanceRating'].mean().reset_index()

# Create a grouped bar chart with one decimal place data labels
plt.figure(figsize=(5, 4))
bars = plt.bar(grouped_data['Gender'], grouped_data['PerformanceRating'], color=['pink', 'skyblue'])
plt.xlabel('Gender')
plt.ylabel('Average Performance Rating')
plt.title('Average Performance Rating by Gender')
plt.ylim(0, 5)  # Set the y-axis limit to match the rating scale (usually 1-5)

# Add data labels above each bar with one decimal place
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.1f}',  # Format the label to one decimal place
                 xy=(bar.get_x() + bar.get_width() / 2, height),  # Position of the label above the bar
                 xytext=(0, 3),  # Offset for the label position
                 textcoords='offset points',
                 ha='center',  # Horizontal alignment
                 va='bottom')  # Vertical alignment

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('/content/WA_Fn-UseC_-HR-Employee-Attrition.csv')



df = pd.DataFrame(data)

# Count the number of employees in each job role
job_role_counts = df['JobRole'].value_counts().reset_index()
job_role_counts.columns = ['JobRole', 'Count']

# Create a horizontal bar graph to visualize job roles and their counts
plt.figure(figsize=(10, 6))
ax = plt.barh(job_role_counts['JobRole'], job_role_counts['Count'], color='skyblue')

plt.xlabel('Count')
plt.ylabel('Job Role')
plt.title('Distribution of Job Roles')
plt.tight_layout()

# Add data labels (counts) to the bars
for i, v in enumerate(job_role_counts['Count']):
    ax[i].set_width(v)
    plt.text(v + 0.2, i, str(v), color='black', va='center', fontsize=10)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
data = pd.read_csv('/content/WA_Fn-UseC_-HR-Employee-Attrition.csv')

# Example: Histogram for 'Age'
plt.figure(figsize=(8, 5))
ax = sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')

# Add value annotations to the histogram
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

plt.show()

# Example: Boxplot for 'MonthlyIncome'
# Calculate statistics
mean_income = data['MonthlyIncome'].mean()
min_income = data['MonthlyIncome'].min()
max_income = data['MonthlyIncome'].max()

# Example: Boxplot for 'MonthlyIncome'
plt.figure(figsize=(10, 5))
sns.boxplot(data=data, y='MonthlyIncome')
plt.title('Monthly Income Distribution')
plt.xlabel('Monthly Income')

# Add value annotations to the boxplot (mean, min, max)
plt.annotate(f'Mean: {mean_income:.2f}', xy=(0.5, mean_income), xytext=(0.5, mean_income + 1000),
             arrowprops=dict(facecolor='blue', shrink=0.05), ha='center', va='bottom')

plt.annotate(f'Min: {min_income}', xy=(0.5, min_income), xytext=(0.5, min_income - 1000),
             arrowprops=dict(facecolor='red', shrink=0.05), ha='center', va='top')

plt.annotate(f'Max: {max_income}', xy=(0.5, max_income), xytext=(0.5, max_income + 1000),
             arrowprops=dict(facecolor='green', shrink=0.05), ha='center', va='bottom')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('/content/WA_Fn-UseC_-HR-Employee-Attrition.csv')

# Create subplots to visualize financial trends
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))

# Financial Trend 1: Monthly Income Distribution
sns.histplot(data=data, x='MonthlyIncome', bins=20, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Monthly Income Distribution')
axes[0, 0].set_xlabel('Monthly Income')

# Add annotations to the first subplot
for p in axes[0, 0].patches:
    axes[0, 0].annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

# Financial Trend 2: Monthly Rate Distribution
sns.histplot(data=data, x='MonthlyRate', bins=20, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Monthly Rate Distribution')
axes[0, 1].set_xlabel('Monthly Rate')
axes[0, 1].set_ylabel('Percentage')

# Add annotations to the second subplot
for p in axes[0, 1].patches:
    axes[0, 1].annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

# Financial Trend 3: Percent Salary Hike Distribution
sns.histplot(data=data, x='PercentSalaryHike', bins=15, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Percent Salary Hike Distribution')
axes[1, 0].set_xlabel('Percent Salary Hike')

# Add annotations to the third subplot
for p in axes[1, 0].patches:
    axes[1, 0].annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

# Financial Trend 4: Monthly Income vs. Monthly Rate
sns.scatterplot(data=data, x='MonthlyIncome', y='MonthlyRate', hue='Attrition', ax=axes[1, 1])
axes[1, 1].set_title('Monthly Income vs. Monthly Rate')
axes[1, 1].set_xlabel('Monthly Income')
axes[1, 1].set_ylabel('Monthly Rate')

# Adjust subplot layout
plt.tight_layout()

# Show the plots
plt.show()
```
